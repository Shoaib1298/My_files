#### Analysis for the system with the noice


```python
import numpy as np
import matplotlib.pyplot as plt

x_0 = 1
gamma = 1
k = 1 
D = 1/2

F_0 = -k*x_0
n=10
sum_a = 0

for j in range(0,n):
    x_t = x_0
    x_time = []
    time = []
    delta_t = 0.000001
    t = 5000
    time_i = 0
    for i in range (0,t):
        random_x = np.random.normal(0, 1)
        if (2*D*delta_t)*random_x != 0 :

            x_t_next = x_t -k*x_t*delta_t + np.sqrt(2*D*delta_t)*random_x
            x_time.append(x_t_next)
            x_t = x_t_next
            time_i = time_i + delta_t
            time.append(time_i)

        else:
            x_t_next = x_t -k*x_t*delta_t 
            x_time.append(x_t_next)
            x_t = x_t_next
            time_i = time_i + delta_t
            time.append(time_i)
    sum_a = sum_a + x_t_next    
    plt.plot(time,x_time)
```
